```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file_loc = r"C:\Users\suhai\Downloads\task_05\RTA Dataset.csv"
df = pd.read_csv(file_loc)
df
```

```python
column_names=df.columns
print(column_names)
```

```python
selected_columns=['Time','Day_of_week','Age_band_of_driver','Sex_of_driver','Driving_experience','Area_accident_occured',
                    'Road_surface_type','Road_surface_conditions','Weather_conditions','Type_of_collision','Cause_of_accident']
df_selected=df[selected_columns]
print(df_selected.head())
```

```python
df_selected['Time']=pd.to_datetime(df_selected['Time'])
df_selected
```

```python
df_selected.isnull().sum()
```

```python
df_selected=df_selected.dropna()
df_selected.isnull().sum()
```

```python
df_selected=df_selected.drop_duplicates()
```

```python
df_selected['Hour']=df_selected['Time'].dt.hour
plt.figure(figsize=(12,6))
sns.countplot(data=df_selected, x='Hour')
plt.title('Distribution of accidents by hour of day')
plt.xlabel('Hour of day')
plt.ylabel('Number of accidents')
plt.show()
```

```python
days_order=['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday']

plt.figure(figsize=(10,6))
sns.countplot(data=df_selected, x='Day_of_week',order=days_order)
plt.title('Distribution of accidents by Day of Week')
plt.xlabel('Day of week')
plt.ylabel('Number of accidents')
plt.show()
```

```python
age_order=['Under 18','18-30','31-50','Over 51','Unknown']
plt.figure(figsize=(10,6))
sns.countplot(data=df_selected, x='Age_band_of_driver', order=age_order)
plt.title('Distribution of accidents by Age band of driver')
plt.xlabel('Age band')
plt.ylabel('Number of accidents')
plt.show()
```

```python
df_selected['Sex_of_driver'].value_counts().plot.pie(autopct='%1.1f%%',startangle=170)
plt.title('Accidents by sex of driver')
plt.show()
```

```python
experience_order=['No Licence','Below 1yr','1-2yr','2-5yr','5-10yr','Above 10yr','unknown']
plt.figure(figsize=(10,6))
sns.countplot(data=df_selected, x='Driving_experience', order=experience_order )
plt.title('Distribution of accidents by driving experience of driver')
plt.xlabel('Driving experience')
plt.ylabel('Number of accidents')
plt.show()
```

```python
column_to_plot='Area_accident_occured'
value_counts=df_selected['Area_accident_occured'].value_counts()
plt.figure(figsize=(12,6))
sns.countplot(data=df_selected, y='Area_accident_occured')
plt.title('Distribution of accidents by Area')
plt.xlabel('Number of accidents')
plt.ylabel('Area')
plt.show()
```

```python
column_to_plot='Road_surface_type'
value_counts=df_selected['Road_surface_type'].value_counts()
plt.figure(figsize=(10,6))
sns.countplot(data=df_selected, y='Road_surface_type')
plt.title('Distribution of accidents by type of road surface')
plt.xlabel('Number of accidents')
plt.ylabel('Type of surface')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.countplot(data=df_selected, x='Road_surface_conditions')
plt.title('Distribution of accidents by road condition')
plt.xlabel('Condition of road')
plt.ylabel('Number of accidents')
plt.show()
```

```python
column_to_plot='Weather_conditions'
value_counts=df_selected['Weather_conditions'].value_counts()
plt.figure(figsize=(10,6))
sns.countplot(data=df_selected, y='Weather_conditions')
plt.title('Distribution of accidents by Weather condition')
plt.xlabel('Number of accidents')
plt.ylabel('Weather conditions')
plt.show()
```

```python
column_to_plot='Cause_of_accident'
value_counts=df_selected['Cause_of_accident'].value_counts()
plt.figure(figsize=(10,6))
sns.countplot(data=df_selected, y='Cause_of_accident')
plt.title('Distribution of accidents by cause of accident')
plt.xlabel('Number of accidents')
plt.ylabel('Cause of accident')
plt.show()
```

```python
# Accidents Heatmap by Hour and Day of the Week
heatmap_data = df_selected.groupby(['Day_of_week', 'Hour']).size().unstack(fill_value=0)
plt.figure(figsize=(14,7))
sns.heatmap(heatmap_data, cmap="YlGnBu", annot=True, fmt="d")
plt.title('Accidents Heatmap by Hour and Day of the Week')
plt.xlabel('Hour of the Day')
plt.ylabel('Day of the Week')
plt.show()
```
